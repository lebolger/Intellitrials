```python

# Import necessary modules
import feedparser
import sqlite3

# RSS Feed URL
rss_url = 'https://clinicaltrials.gov/api/rss?firstPost=2020-01-01_&lastUpdPost=2023-01-01_&aggFilters=funderType%3Aindustry+nih+fed%2CstudyType%3Aint+exp+exp_inter+exp_indiv+exp_treat&dateField=LastUpdatePostDate'

# Parse RSS Feed
feed = feedparser.parse(rss_url)

# Connect DB
conn = sqlite3.connect('clinicaltrials_rss.db')
cursor = conn.cursor()

# Create Table
cursor.execute('''
CREATE TABLE IF NOT EXISTS trials (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT,
    link TEXT,
    published TEXT,
    summary TEXT
)
''')

# Insert Data
for entry in feed.entries:
    cursor.execute('''
    INSERT INTO trials (title, link, published, summary)
    VALUES (?, ?, ?, ?)
    ''', (entry.title, entry.link, entry.published, entry.summary))

conn.commit()
conn.close()

```

```python
# create data frame
import pandas as pd
# Connect DB again to read data into DataFrame
conn = sqlite3.connect('clinicaltrials_rss.db')
df = pd.read_sql_query('SELECT * FROM trials', conn)
conn.close()
# Display the DataFrame
print(df)
```

```python
# Count rows and columns
num_rows = df.shape[0]  # Number of rows
num_columns = df.shape[1]  # Number of columns

# Print results
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```
